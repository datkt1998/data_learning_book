```python
#Generate a dummy dataset.
X = np.random.randint(1,200,100).reshape(20,5)
print(X)

# mean Centering the data  
X_mean = X - np.mean(X , axis = 0)

# Calculate the Covariance Matrix
cov_mat = np.cov(X_mean , rowvar = False)

#Calculating Eigenvalues and Eigenvectors of the covariance matrix
eigen_values , eigen_vectors = np.linalg.eigh(cov_mat)

#sort the eigenvalues in descending order
sorted_index = np.argsort(eigen_values)[::-1]
sorted_eigenvalue = eigen_values[sorted_index]
sorted_eigenvectors = eigen_vectors[:,sorted_index] #similarly sort the eigenvectors 

# select the first n eigenvectors, n is desired dimension of our final reduced data.
n_components = 4 #you can select any number of components.
eigenvector_subset = sorted_eigenvectors[:,0:n_components]

#Transform the data 
X_reduced = np.dot(eigenvector_subset.transpose(),X_mean.transpose()).transpose()
```

```python
# correlation (X, Y): scatter-plot Y vs each X variable
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from plotly.subplots import make_subplots
import plotly.express as px
import plotly.graph_objects as go

def scatter_plot(X, Y = None, X_name = None, Y_name = None):
    if Y is None:
        def corrfunc(x, y, **kwds):
            cmap = kwds['cmap']
            norm = kwds['norm']
            ax = plt.gca()
            ax.tick_params(bottom=False, top=False, left=False, right=False)
            sns.despine(ax=ax, bottom=True, top=True, left=True, right=True)
            r, _ = pearsonr(x, y)
            facecolor = cmap(norm(r))
            ax.set_facecolor(facecolor)
            lightness = (max(facecolor[:3]) + min(facecolor[:3]) ) / 2
            ax.annotate(f"r={r:.2f}", xy=(.5, .5), xycoords=ax.transAxes,
                        color='white' if lightness < 0.7 else 'black', size=26, ha='center', va='center')

        # plot correlation
        df = pd.DataFrame(X, columns = X_name)
        g = sns.pairplot(df, diag_kind="kde")
        g.map_upper(corrfunc, cmap=plt.get_cmap('seismic'), norm=plt.Normalize(vmin=-.5, vmax=.5))
        g.fig.subplots_adjust(wspace=0.06, hspace=0.06) # equal spacing in both directions
        plt.show()
    
    else:
        fig = make_subplots(rows=3, cols=5,)
        fig_data = tuple()
        for index, feature_name in enumerate(X_name):
            df = pd.DataFrame({feature_name:X[:,index] , Y_name:Y})
            fig.add_trace(go.Scatter(x=df[feature_name], y=df[Y_name], mode='markers', name = feature_name) ,    row=index//5+1,    col=index%5+1, )
        fig.update_layout(autosize=False,    width=1200,    height=800,)
        fig.show()

scatter_plot(X = boston_data.data, Y = boston_data.target, X_name = boston_data.feature_names, Y_name = 'HOUSE_PRICE' )
```

```python
# Pre-processing the Data:
boston_df = pd.DataFrame(boston_data.data)
boston_df.columns = boston_data.feature_names
boston_df_target = np.asarray(boston_data.target)
boston_df['HOUSE_PRICE'] = pd.Series(boston_df_target)
X = boston_df.iloc[:, :-1]
Y = boston_df.iloc[:, -1]
boston_df.head(3)
```

# Singular Value Decomposition (SVD)

## Continuous dataset

```python
# load dataset
import pandas as pd
import sklearn.datasets as data

iris = data.load_iris()
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris['target'] = iris.target
print(df_iris.shape)
df_iris.head(3)
```

```python
import numpy as np
X = df_iris.loc[:,df_iris.columns[:-1]]
X_demean = X - np.mean(X, axis=0)
U, S, Vt = np.linalg.svd(X_demean)
```

```python
np.power(S,2)
```

```python
# the left-singular vectors of  𝑋 , where columns are the set of orthonormal eigenvectors of  𝑋𝑋⊺
U.shape # equal len of data
```

```python
# the right-singular vectors  𝑋 , where the columns are the set of orthonormal eigenvectors of  𝑋⊺𝑋
Vt.shape # equal number of columns
```

```python
# singular value of data
# diagonal matrix  Σ  with entries  𝜎∈ℝ  that are the non-negative singular values of  𝑋
# S là vector đường chéo của matrix (150,4), 
# trong đó dòng từ 5 tới 150 là matrix 0 nên suy ra cột từ 5 tới 150 của U không có ý nghĩa
S.shape
```

```python
cum_var_explained = np.cumsum(np.square(S))/np.sum(np.square(S))
cum_var_explained
```

```python
# plot number singular value and cum_explained
import plotly.express as px

fig = px.line(x=range(1,5), y = cum_var_explained,range_y = [0.9,1.05], range_x = [0,5])
fig.show()
```

```python
import plotly.graph_objects as go
from sklearn.preprocessing import normalize
fig = go.Figure()
fig.add_trace(go.Scatter(x=[1,2,3,4], y=normalize(S.reshape((1,-1)))[0],  mode='markers', name='singular values (normalized)'))
fig.add_trace(go.Scatter(x=[1,2,3,4], y=cum_var_explained.round(3),  mode='lines', name='cumulative variance explained'))

fig.update_layout(autosize=False,    width=700,    height=500,)
fig.show()
```

### Check the Multi-collinearity

There are various techniques to find the presence of multicollinearity in the data, some of them are:

- Getting very high standard errors for regression coefficients
- The overall model is significant, but none of the coefficients are significant
- Large changes in coefficients when adding predictors
- High Variance Inflation Factor (VIF) and Low Tolerance

```python
contingency = pd.crosstab(df_countries.country, df_countries.language, margins=False) # pivottable with count
contingency.index = ['Canada', 'USA', 'England', 'Italy', 'Switzerland'] # country
contingency.columns = ['English', 'French', 'Spanish', 'German', 'Italian'] # language
contingency
```

```python
prob = contingency/contingency.sum().sum()
prob
```

```python
Mu= prob.sum(axis=1).values.reshape(-1,1)@prob.sum(axis=0).values.reshape(1,-1)
Lambda = (prob - Mu) / np.sqrt(Mu)
Lambda
```

```python
from scipy.sparse.linalg import svds
```

```python
def svd_process(train, k):
    utilMat = np.array(train) if type(train) != np.array else train
    
    # DE-MEAN process
    # the nan or unavailable entries are masked
    mask = np.isnan(utilMat)
    masked_arr = np.ma.masked_array(utilMat, mask)
    item_means = np.mean(masked_arr, axis=0)
    # nan entries will replaced by the average rating for each item
    utilMat = masked_arr.filled(item_means)
    x = np.tile(item_means, (utilMat.shape[0],1)) # the per item average from all entries
    utilMat = utilMat - x # Remove x (mean), the above mentioned nan entries will be essentially 0 now
    
    # SVD: U and V are user and item features
    U, s, V=svds(utilMat, k)
    s=np.diag(s) # convert singular vector to Diagonalizable matrix
    
    # estimate the apporoximate matrix
    s_root=sqrtm(s) # then s = s_root.dot(s_root)
    Usk=np.dot(U,s_root)
    skV=np.dot(s_root,V)
    UsV = np.dot(Usk, skV)
    UsV = UsV + x
    # print("svd done")
    return UsV
```

```python
usv = svd_process(ratings_matrix, 50)
x = (abs(usv.round()-ratings_matrix))
unique, counts = np.unique(x, return_counts=True)
print(np.asarray((unique, counts)).T.astype(int))
```

```python
def rmse(true, pred):
    # this will be used towards the end
    x = true - pred
    return sum([xi*xi for xi in x])/len(x)

# to test the performance over a different number of features
no_of_features = [8,10,12,14,17,30,50,100]
train2 =train.copy()
train2['rating']=np.nan
total_ratings = pd.concat([train2,test]).pivot(index = 'user_id', columns = 'movie_id', values = 'rating')\
.reindex(user_id_idx).reindex(movie_id_idx,axis=1).values
for f in no_of_features: 
    svdout = svd_process(ratings_matrix, k=f)
    df_pred = pd.DataFrame(np.where(np.isnan(total_ratings),np.nan,svdout),user_id_idx,movie_id_idx)
    df_pred = df_pred.reset_index().melt(id_vars='index',value_vars=df_pred.columns,var_name='movie_id',value_name='rating_pred',)\
    .dropna().rename(columns = {'index':'user_id'})
    check = test.merge(df_pred,how='left').dropna()
    print('no_of_features = ',f,',RMSE = ',rmse(check['rating'], check['rating_pred']))
```
